```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## Membuat diagram dengan Marker Value

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='D')
plt.show()
```

## Membuat Diagram dengan Format Marker-Line-Color

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='D', linestyle='-', c='g')
plt.show()
```

```python
# Shorthand
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'D-g')
plt.show()
```

```python
# Shorthand
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'D--g')
plt.show()
```

```python
# Shorthand
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'D:g')
plt.show()
```

## Dengan Marker Parameter

```python
# Shorthand
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='D', ms='12', mec='red', mfc='#4caf50')
plt.show()
```

## Membuat Diagram dengan Title dan Label

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='D', mec='red', mfc='#4caf50')

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
font_title = {'family': 'serif', 'color': 'blue', 'size': 20}
font_label = {'family': 'serif', 'color': 'darkred', 'size': 15}

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='D', mec='red', mfc='#4caf50')

plt.title("Sports Watch Data", fontdict=font_title)
plt.xlabel("Average Pulse", fontdict=font_label)
plt.ylabel("Calorie Burnage", fontdict=font_label)

plt.show()
```

## Diagram dengan grid

```python
font_title = {'family': 'serif', 'color': 'blue', 'size': 20}
font_label = {'family': 'serif', 'color': 'darkred', 'size': 15}

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='D', mec='red', mfc='#4caf50')

plt.title("Sports Watch Data", fontdict=font_title)
plt.xlabel("Average Pulse", fontdict=font_label)
plt.ylabel("Calorie Burnage", fontdict=font_label)

plt.grid()

plt.show()
```

```python
font_title = {'family': 'serif', 'color': 'blue', 'size': 20}
font_label = {'family': 'serif', 'color': 'darkred', 'size': 15}

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='D', mec='red', mfc='#4caf50')

plt.title("Sports Watch Data", fontdict=font_title)
plt.xlabel("Average Pulse", fontdict=font_label)
plt.ylabel("Calorie Burnage", fontdict=font_label)

plt.grid(axis='y', color='green')

plt.show()
```

```python
font_title = {'family': 'serif', 'color': 'blue', 'size': 20}
font_label = {'family': 'serif', 'color': 'darkred', 'size': 15}

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='D', mec='red', mfc='#4caf50')

plt.title("Sports Watch Data", fontdict=font_title)
plt.xlabel("Average Pulse", fontdict=font_label)
plt.ylabel("Calorie Burnage", fontdict=font_label)

plt.grid(axis='x', linestyle='--')

plt.show()
```
