## Membuat Scatter Plot

```python
x = np.array([5, 8, 7, 2, 17, 2, 9, 4, 15, 12, 9, 6])
y = np.array([55, 58, 47, 22, 174, 25, 92, 34, 17, 72, 92, 63])

plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
x = np.array([5, 8, 7, 2, 17, 2, 9, 4, 15, 12, 9, 6])
y = np.array([55, 58, 47, 22, 174, 25, 92, 34, 17, 72, 92, 63])

plt.scatter(x, y)

x = np.array([5, 4, 21, 6, 3, 8, 51, 56, 15, 12, 97, 67])
y = np.array([51, 53, 47, 32, 174, 25, 92, 34, 17, 72, 92, 63])

plt.scatter(x, y)


plt.grid()
plt.show()
```

```python
x = np.array([5, 8, 7, 2, 17, 2, 9, 4, 15, 12, 9, 6])
y = np.array([55, 58, 47, 22, 174, 25, 92, 34, 17, 72, 92, 63])

plt.scatter(x, y, color="purple")

x = np.array([5, 4, 21, 6, 3, 8, 51, 56, 15, 12, 97, 67])
y = np.array([51, 53, 47, 32, 174, 25, 92, 34, 17, 72, 92, 63])

plt.scatter(x, y,color="red")


plt.grid()
plt.show()
```

```python
x = np.array([5, 8, 7, 2, 17, 2, 9, 4, 15, 12, 9, 6])
y = np.array([55, 58, 47, 22, 174, 25, 92, 34, 17, 72, 92, 63])
size = np.array([10, 25, 50, 100, 250, 500, 30, 45, 5, 150, 300, 100])

plt.scatter(x, y, s=size, color="orange")

plt.grid()
plt.show()
```

## Membuat Bar Chart

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width=0.2, color="#4CAF50")
plt.show()
```

## Menampilkan Pie Chart

```python
x = np.array([35, 25, 25, 15])

plt.pie(x)
plt.show()
```

```python
colors = ["cyan", "hotpink", "b", "purple"]
x = np.array([35, 25, 25, 15])

plt.pie(x, colors=colors)
plt.show()
```

```python
labels = ["Apples", "Bananas", "Cherries", "Dates"]
explode = [0, 0, 0.1, 0]
x = np.array([35, 25, 25, 15])

plt.pie(x, labels=labels, explode=explode)
plt.legend(loc="upper left")
plt.show()
```
