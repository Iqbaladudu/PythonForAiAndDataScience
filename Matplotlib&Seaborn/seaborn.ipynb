```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

## Membuat Diagram Sederhana Dengan Seaborn

```python
data = sns.load_dataset("iris")
```

```python
data
```

```python
np.unique(data["species"])
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
```

## Membuat Diagram Sederhana Dengan Seaborn dan Matplotlib

```python
plt.figure(figsize=(10, 12))

sns.set_palette('vlag')
plt.subplot(211)
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Title Using Matplotlib Function")
plt.plot()

sns.set_palette('Accent')
plt.subplot(212)
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Title Using Matplotlib Function")
plt.plot()

plt.show()
```

## Membuat Scatter Plot

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

## Membuat Bar Chart

```python
sns.barplot(x='species', y='sepal_length', data=data)
plt.show()
```

```python
sns.countplot(x='species', data=data)
plt.show()
```

## Membuat Histogram

```python
sns.histplot(data['sepal_width'])
plt.show()
```

## Membuat Box Plot

```python
sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

## Violin Plot

```python
sns.violinplot(x="species", y="sepal_width", data=data)
```

## Pair Plot

```python
# hue => fitur yang menjadi acuan
sns.pairplot(data=data, hue='species')
plt.show()
```
