## Membuat Dua Plot Secara Horizontal

```python
# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

# 1 row 2 color dan diisi kolom ke pertama
plt.subplot(1, 2, 1)
plt.plot(x, y)

# Plot 2

x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

# 1 row 2 color dan diisi di kolom ke dua
plt.subplot(1, 2, 2)
plt.plot(x, y)

plt.show()
```

## Memberi Title Pada Dua Plot

```python
# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

# 1 row 2 color dan diisi kolom ke pertama
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('SALES')

# Plot 2

x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

# 1 row 2 color dan diisi di kolom ke dua
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('INCOME')

plt.show()
```

## Memberi Dua Title Utama Pada Dua Plot

```python
plt.figure(figsize=(12, 4))

# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

# 1 row 2 color dan diisi kolom ke pertama
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('USERS')

# Plot 2

x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

# 1 row 2 color dan diisi di kolom ke dua
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('INCOME')

plt.suptitle("TRACTION JUNE 2022")
plt.show()
```
